```python
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import accuracy_score
import os
```

```python
directory = "data/SMNI_CMI_TRAIN/Train/"
filenames = os.listdir(directory)
all_frames = []
for name in filenames:
    temp = pd.read_csv(directory+name, header=0)
    all_frames.append(temp)

# all_data = pd.concat(all_frames)
# all_data.head(5)
# all_frames.head()
```

```python
# S1 obj: one picture
# S2 match: two identical pictures
# S2 nomatch: two different pictures
# all_data.nunique()
# all_frames.nunique()
all_frames[0].columns
```

```python
def makeOneRow(df):
    columns = ["trial number", "subject identifier", "name", "matching condition"]
    values = {}
    for sensorPos, sampleNum, sensorVal in zip(df["sensor position"], df["sample num"], df["sensor value"]):
        values[sensorPos+'-'+str(sampleNum)] = [sensorVal]
    for col in columns:
        if col == "matching condition":
            if df[col][0] == "S2 nomatch,":
                values[col] = ["S2 nomatch"]
            else:
                values[col] = [df[col][0]]
        elif col == "subject identifier":
            if df[col][0] == "a":
                values[col] = ["1"]
            else:
                values[col] = ["0"]
        else: 
            values[col] = [df[col][0]]
    return pd.DataFrame(values)  
            
```

```python
bigData = pd.concat([makeOneRow(df) for df in all_frames])
```

```python
testDir = "data/SMNI_CMI_Test/Test/"
testFiles = os.listdir(testDir)
all_test = []
for name in testFiles:
    temp = pd.read_csv(testDir+name, header=0)
    all_test.append(temp)
bigTest = pd.concat([makeOneRow(df) for df in all_test])
bigTest.head()
```

```python
def Model(df_train, matchCond, dropCols):
    temp = df_train[(df_train["matching condition"] == matchCond)]
    X = temp.drop(dropCols, axis=1)
    y = temp["subject identifier"]
    model = MLPClassifier(hidden_layer_sizes=(256, 64, 8), activation='relu', solver='adam', max_iter=1000)
    return model.fit(X, y)

def testModel(model, df_test, matchCond, dropCols):
    temp = df_test[(df_test["matching condition"] == matchCond)]
    X = temp.drop(dropCols, axis=1)
    y = temp["subject identifier"]
    return accuracy_score(model.predict(X), y)
```

```python
dropColumns = ["trial number", "matching condition", "name"]
models = {}
for cond in bigData["matching condition"].unique():
    models[cond] = Model(bigData, cond, dropColumns)
```

```python
print("All Sensors Accuracy")
for cond, model in models.items():
    print("{}: {}".format(cond, testModel(model, bigTest, cond, dropColumns)))
```

```python
def ModelPosition(df_train, matchCond, sensorPos):
    temp = df_train[(df_train["matching condition"] == matchCond)]
    posColumns = [sensorPos+'-'+str(i) for i in range(0, 256)]
    X = temp[posColumns]
#     print(X.dtypes)
    y = temp["subject identifier"]
    model = MLPClassifier(hidden_layer_sizes=(256, 64, 8), activation='relu', solver='adam', max_iter=1000)
    return model.fit(X, y)

def testPosModel(model, df_test, matchCond, sensorPos, dropCols):
    temp = df_test[(df_test["matching condition"] == matchCond)]
    posColumns = [sensorPos+'-'+str(i) for i in range(0, 256)]
    X = temp[posColumns]
    y = temp["subject identifier"]
    return accuracy_score(model.predict(X), y)
```

```python
dropColumns = ["trial number", "name"]
sensorModels = {}
for cond in bigData["matching condition"].unique():
    sensorModels[cond] = {}
    for pos in all_frames[0]["sensor position"].unique():
        sensorModels[cond][pos] = ModelPosition(bigData, cond, pos)
pd.DataFrame(sensorModels)
```

```python
print("Sensor Postions Accuracy")
# {cond: pos: model}
posScore = {}
for cond in sensorModels:
    print(cond)
    posScore[cond] = {}
    for i, (pos, model) in enumerate(sensorModels[cond].items()):
        result = testPosModel(model, bigTest, cond, pos, dropColumns)
        posScore[cond][pos] = result
        print("{}: {}".format(pos, result), end='\t')
        if i % 8 == 1:
            print()
```

```python
for cond in posScore:
    for pos, score in posScore[cond].items():
        if score > 0.75:
            print("{}: {}".format(pos, score))
```

```python
for cond in posScore:
    for pos, score in posScore[cond].items():
        if score < 0.5:
            print("{}: {}".format(pos, score))
```

```python
def testPosTrials(posModels, trialData):
    predictions = []
    for pos, model in posModels[trialData["matching condition"]].items():
        posColumns = [pos+'-'+str(i) for i in range(0, 256)]
        predictions.append(model.predict([trialData[posColumns]]))
    predictions = [int(val[0]) for val in predictions]
    return sum(predictions)/len(predictions)
```

```python
trialScores = {}
for index, row in bigTest.iterrows():
    trialScores[row["trial number"]] = {}
    trialScores[row["trial number"]]["expected"] = row["subject identifier"]
    trialScores[row["trial number"]]["actual"] = testPosTrials(sensorModels, row)
```

```python
for trial in bigTest["trial number"].unique():
    print("{}: {}, {}".format(trial, trialScores[trial]["expected"], round(trialScores[trial]["actual"])))
```
